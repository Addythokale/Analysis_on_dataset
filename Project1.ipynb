```python
import pandas as pd
```

```python
data=pd.read_csv("houseprice.csv")
```

```python
data.dropna(inplace=True)
```

```python
data
```

### 1) Find the dealer who has under construction & RERA approved with 2BHK flat but area of that flat should be between 900 to 1300 sq.feet 

```python
a1=data.loc[(data["POSTED_BY"]=="Dealer")&(data["UNDER_CONSTRUCTION"]==1) & (data["RERA"]==1) & (data["BHK_NO."]==2) &
            (data["SQUARE_FT"]>=900) & (data["SQUARE_FT"]<=1300)]
```

```python
a1
```

### 2) Find those flats those are ready to move but not approved by RERA and cost of that flat that should be between 50 to 75 lakhs.

```python
b1=data.loc[(data["READY_TO_MOVE"]==1) & (data["RERA"]==0) & (data["TARGET(PRICE_IN_LACS)"]>=50) & 
            (data["TARGET(PRICE_IN_LACS)"]<=75)]
```

```python
b1
```

### 3) Find those owners and dealers who has 2BHK & 3BHK flats and area of that flat is less than 1000 sq.feet also cost is greater than 45 lakh. 

```python
c1=data.loc[((data["POSTED_BY"]=="Owner") | (data["POSTED_BY"]=="Dealer")) & ((data["BHK_NO."]==2) | (data["BHK_NO."]==3)) & 
            (data["SQUARE_FT"]<=1000) & (data["TARGET(PRICE_IN_LACS)"]>=45)]
```

```python
c1
```

### 4) Find those Owners and Dealers who has RERA approved with 2BHK and 3BHK flats then incerement their target price by 5 lakhs.

```python
d1=data.loc[((data["POSTED_BY"]=="Owner") | (data["POSTED_BY"]=="Dealer")) & (data["RERA"]==1) & ((data["BHK_NO."]==2) | (data["BHK_NO."]==3))]
```

```python
d1
```

```python
With_Increment=[]
for x in d1["TARGET(PRICE_IN_LACS)"]:
    x=x+5
    With_Increment.append(x)
With_Increment
```

```python
d1["New_Target"]=With_Increment
```
